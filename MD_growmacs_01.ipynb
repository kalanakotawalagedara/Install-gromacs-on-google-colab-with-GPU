```python
# install required dependencies
!apt-get update
!apt-get install -y cmake build-essential libfftw3-dev

# Download GROMACS 2024.4 source
!wget https://ftp.gromacs.org/pub/gromacs/gromacs-2024.4.tar.gz

# Extract the tar file
!tar xvf gromacs-2024.4.tar.gz

# Create build directory
!mkdir gromacs-2024.4/build
%cd gromacs-2024.4/build

# Configure with CMake
!cmake .. -DGMX_GPU=ON -DGMX_GPU=CUDA -DGMX_BUILD_OWN_FFTW=ON -DREGRESSIONTEST_DOWNLOAD=ON

# Build GROMACS (using 2 cores, adjust if needed)
!make -j 2

# Install GROMACS
!make install

# Source GROMACS environment (needed after installation)
!source /usr/local/gromacs/bin/GMXRC
```
